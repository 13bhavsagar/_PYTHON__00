1.What is the difference between a function and a method in Python?
>A function is an independent block of code defined using def that can be called anywhere in the program.

>A method is a function that belongs to an object and is called using object.method().

2.Explain the concept of function arguments and parameters in Python?
>Parameters are variables listed in a function’s definition. They act as placeholders for values.

>Arguments are actual values passed to a function when it is called.

3.What are the different ways to define and call a function in Python?
>Regular Function: Defined using def, called with arguments.

>Lambda Function: Anonymous, single-expression function.

>Recursive Function: Calls itself for repetitive tasks.

4.What is the purpose of the `return` statement in a Python function?
>The return statement in a Python function is used to send a value back to the caller and terminate the function’s execution.

5.What are iterators in Python and how do they differ from iterables?
>Iterators: Objects that produce values one at a time using __iter__() and __next__(). They do not store all values in memory.

>Iterables: Objects that can be looped over (e.g., lists, tuples, strings). They implement the __iter__() method.

6.Explain the concept of generators in Python and how they are defined.
>A generator is a special type of iterator that generates values lazily (one at a time) using the yield keyword, making it memory-efficient.

7.What are the advantages of using generators over regular functions?
>Memory Efficient – Generates values one at a time without storing them in memory.

>Faster Execution – Produces results on demand, avoiding unnecessary computations.

>State Retention – Uses yield to pause and resume execution instead of restarting.

>Handles Infinite Sequences – Can stream data infinitely without memory overflow.

>No List Creation – Unlike regular functions, generators don’t create and store entire lists

8.What is a lambda function in Python and when is it typically used?
>A lambda function is an anonymous, single-expression function defined using the lambda keyword.

>Typical Uses of Lambda Functions:
Short, Simple Functions – Used for quick operations without defining a full def function.

In map(), filter(), sorted() – Often used for applying logic inline.

9.Explain the purpose and usage of the `map()` function in Python.
>The map() function applies a given function to each item in an iterable (like a list or tuple) and returns an iterator of the results. It’s used for transforming data in a concise and efficient way.


10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
>map() applies a function to each item in an iterable, returning an iterator of results.

>reduce() (from functools) applies a binary function cumulatively to reduce an iterable to a single value.

>filter() filters elements from an iterable based on a condition, returning only those that satisfy it.

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciSeries:
    def __init__(self,n):
        self.n=n
        self.a=0
        self.b=1
        self.count=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count>self.n:
            raise StopIteration
        if self.count==0:
            self.count+=1
            return self.a
        elif self.count==1:
            self.count+=1
            return self.b
        else :
            fib=self.a+self.b
            self.a,self.b=self.b,fib
            self.count+=1
            return fib
n=int (input("enter the number"))
fibo=FibonacciSeries(n)
for i in fibo:
    print(i,end="  ")
```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent.
def  power(x):
    for i in range(x+1):
        yield 2**i
x=int(input("enter the value  of x:"))

p=power(x)
for i in p:
    print(i,end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    with open(file_path,'r') as f:
        for i in f:
            yield i.strip()
file_path="file.txt"
for i in read_file_line_by_line(file_path):
    print(i)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
st=input("enter the string")

a=list(filter(lambda x:x if x not in ['a','e','i','o','u'] else "",st))
print("list form:",a)
print("string form:","".join(a))
```
