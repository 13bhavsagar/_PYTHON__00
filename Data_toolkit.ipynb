#**THEORATICAL QUESTIONS**

1.What is NumPy, and why is it widely used in Python?
>NumPy (Numerical Python) is a powerful open-source Python library used primarily for numerical computations and handling large multi-dimensional arrays and matrices. It's one of the foundational libraries in the Python data science and scientific computing ecosystem.

2.How does broadcasting work in NumPy?
>Broadcasting describes how NumPy automatically expands arrays of different shapes so that arithmetic operations (like +, -, *, /) can be performed element-wise.

3.What is a Pandas DataFrame?
>A Pandas DataFrame is a 2D labeled data structure like a table with rows and columns.
Each column can have a different data type (int, float, string, etc.).
It's widely used for data analysis, manipulation, and handling real-world datasets.
You can filter, sort, group, and analyze data easily with built-in functions.

4.Explain the use of the groupby() method in Pandas?
>The groupby() method in Pandas is used to group data based on one or more columns, allowing you to perform aggregate operations (like sum, mean, count) on each group.

5.Why is Seaborn preferred for statistical visualizations?
>Seaborn is preferred for statistical visualizations because it makes beautiful and informative plots with minimal code.
It has built-in support for complex statistical plots like box plots, violin plots, and heatmaps.
Seaborn integrates well with pandas DataFrames, making it easy to plot directly from datasets.
It automatically handles data aggregation and adds helpful visual features like confidence intervals.

6.What are the differences between NumPy arrays and Python lists?
>Performance: NumPy arrays are much faster and more memory-efficient than Python lists for large data.

>Data Types: NumPy arrays store elements of the same data type, while Python lists can hold mixed types.

>Functionality: NumPy supports advanced mathematical operations (e.g., vectorization, broadcasting), which lists do not.

>Shape & Dimensions: NumPy supports multi-dimensional arrays (e.g., 2D, 3D), while lists require nesting.

>Built-in Operations: NumPy allows element-wise operations out-of-the-box; Python lists require loops.

7.What is a heatmap, and when should it be used?
>A heatmap is a color-coded matrix used to visualize data values.
It helps quickly spot patterns, correlations, or outliers in large datasets.
Often used for correlation matrices, activity tracking, or missing data.
Use it when you need a visual summary of complex numerical data.

8.What does the term “vectorized operation” mean in NumPy?
>In NumPy, a vectorized operation refers to performing operations on entire arrays (or large chunks of data) without the need for explicit loops.
NumPy achieves this by using highly optimized C-based implementations under the hood, making the operation much faster than Python’s native looping over individual elements.

10.What is the significance of hierarchical indexing in Pandas?

17.Explain the difference between apply() and map() in Pandas?
>In Pandas, both apply() and map() are used to apply functions to data, but they differ in how they work and the types of objects they are applied to.

>Function Application:apply(): Can be used on both Series and DataFrame. It applies a function along either axis (rows or columns) of the DataFrame or to each element of a Series.map(): Specifically used on Series to apply a function element-wise (i.e., to each value of the Series).

>Data Type Handling:apply(): More flexible; it can apply any function (including complex functions) to the entire row or column in a DataFrame or element-wise to a Series.map(): Typically used for simpler, element-wise transformations on a Series (e.g., replacing values, applying a function, or mapping values via a dictionary).

>Return Type:apply(): When used on a DataFrame, it can return a Series, a DataFrame, or a scalar depending on the applied function. For a Series, it returns a Series with transformed values.map(): Always returns a Series, with each element replaced by the mapped result.

>Performance:apply(): May be slower than map() when applied to a Series, as it is more flexible and can apply functions to entire rows or columns in a DataFrame.map(): Tends to be faster when used on a Series for simple transformations, especially with dictionary mappings.

Additionally, NumPy's random sampling and memory efficiency support handling large datasets, while ufuncs ensure fast, element-wise operations.

19.How does Pandas simplify time series analysis?

20.What is the role of a pivot table in Pandas?
>A pivot table in Pandas is a powerful data manipulation tool used to summarize and aggregate data, particularly when working with large datasets. It allows you to reorganize data into a more meaningful and readable format by grouping data based on one or more columns, and then applying aggregate functions like sum, mean, count, etc.

21.Why is NumPy’s array slicing faster than Python’s list slicing?
>NumPy's array slicing is faster than Python's list slicing because NumPy arrays are designed for efficient numerical computations, utilizing low-level C implementations that optimize memory access and data manipulation.

22.What are some common use cases for Seaborn?

It is highly effective for visualizing categorical data with plots like barplots, boxplots, and violin plots. Seaborn also excels in creating statistical visualizations, such as regression plots and heatmaps for correlation matrices. The library supports multi-plot grids for comparing different subsets of data, and it allows for easy customization of aesthetics, making it ideal for publication-quality visuals. These features make Seaborn a go-to tool for data visualization in Python.
